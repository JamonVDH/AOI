```python
import os, osgeo, math, Image, datetime, time, numpy as np
from osgeo import gdal, ogr, osr
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.path as mpath
import matplotlib.lines as mlines
import matplotlib.patches as mpatches
from matplotlib.collections import PatchCollection
%matplotlib inline
%pylab inline
start_time = time.time()

root = 'C:/Git-projects/AOI/sample_data/out-polygon_overlays/'
#root = 'C:/Users/Jamon/Google Drive/projects_shared/Brian--village_AOI/code/sampledata/polygonOverlays/' # dir with all poly shp's to plot
image_dir = root # dir to hold output plots
samplePoly = True # randomly sample polygons?
samplePerc = 0.01 # % percent sample of polygons; ignore if samplePoly = False
area_thresh = 0 # minimum area polygon to plot; set to 0 if not relevant
simplify_geom = False # only plot poly convex hull?
fillColor = True # fill polygons with color?
lineColor = 'black'
plot_dpi = 300
```

```python
def plotFig (figName, plot_dimx, plot_dimy):
    fig = matplotlib.pyplot.gcf()
    fig.set_size_inches(plot_dimx,plot_dimy)
    plt.savefig(image_dir+figName+'.png',dpi=plot_dpi, bbox_inches='tight', pad_inches=0.1)
    Image.open(image_dir+figName+'.png').save(image_dir+figName+'.jpg','JPEG')
    os.remove(image_dir+figName+'.png')

def polyOverlap(polyFeat):                    
    geom = polyFeat.GetGeometryRef() 
    area = geom.ConvexHull().Area()
    if area > area_thresh:
        x_cent = geom.Centroid().GetX() # centroids act as reference (x,y) for shift and overlap
        y_cent = geom.Centroid().GetY()
        if simplify_geom: geomWkt = geom.ConvexHull().ExportToWkt()
        else: geomWkt = geom.ExportToWkt()
        geomWktList = geomWkt.split(',')[1:] # skip first node since it's encountered in last entry
        
        path_data = []
        for i,each in enumerate(geomWktList):
            numVertices = len(geomWktList)
            latlon = each.split(' ')
            x_shift=shift-x_cent # distance from arbitrary pt and centroid
            y_shift=shift-y_cent
            x = float(latlon[0].strip('('))+x_shift # shifted x value
            y = float(latlon[1].strip(')'))+y_shift
            xlist.append(x)
            ylist.append(y)
            if i == (numVertices-1): # if last point, open/close using this x,y
                path_data.append((79,[x,y])) # 79: Path.CLOSEPOLY
                path_data.insert(0,(1,[x,y])) # 1: Path.MOVETO
            else: path_data.append((2,[x,y])) # 2: Path.LINETO
        codes,verts = zip(*path_data)
        path = mpath.Path(verts,codes)
        return mpatches.PathPatch(path)
    else:
        return None
```

```python
# http://matplotlib.org/examples/shapes_and_collections/artist_reference.html
drv = ogr.GetDriverByName('ESRI Shapefile')

for fileShp in os.listdir(root):
    if fileShp.endswith(".shp"): # consider all shp's in directory
        shp = drv.Open(root+fileShp)
        layer = shp.GetLayer(0)
        polyCount = layer.GetFeatureCount()
        srs = layer.GetSpatialRef()
        #x_min, x_max, y_min, y_max = layer.GetExtent()
        
        fig, ax = plt.subplots()
        patches=[]
        xmin,xmax,ymin,ymax=0,0,0,0 # values to be updated
        shift = 1000000 # arbitrary big number to shift poly centroids
        xlist=[] # list to hold all x's to find max/min x value for plot limits
        ylist=[]
        sample = np.random.random_integers(0,polyCount,int(samplePerc*polyCount))
        sample = sample.tolist()
        j=0 # poly feature counter
        used=0
        for polyFeat in layer:
            if samplePoly:
                if j not in sample:
                    j+=1
                else:
                    patch = polyOverlap(polyFeat)
                    if patch != None:
                        patches.append(patch)
                        used+=1
                    j+=1
            else:
                patch = polyOverlap(polyFeat)
                if patch !=None:
                    patches.append(patch)
                    used+=1
                j+=1

        # https://stackoverflow.com/questions/6028675/setting-color-range-in-matplotlib-patchcollection?rq=1
        colors = 100*np.random.random(100)
        # http://matplotlib.org/examples/color/colormaps_reference.html
        if fillColor:
            cmap = matplotlib.cm.autumn
            collection = PatchCollection(patches,alpha=.1,cmap=cmap,color=lineColor)
            collection.set_clim([90, 100])
        else:
            cmap = matplotlib.cm.gray
            collection = PatchCollection(patches,alpha=.1,cmap=cmap,color=lineColor)
            collection.set_clim([0, 0.1])
        collection.set_array(np.array(colors))
        ax.add_collection(collection)
        #plt.colorbar(collection)
        plt.xlim(min(xlist),max(xlist))
        plt.ylim(min(ylist),max(ylist))
        plt.axis('off')
        plt.tight_layout()
        
        plotFig(fileShp.strip('.shp'),6,6)
        print '     total polygons: '+str(j-1)
        print '     overlapping polygons: '+str(used)
        
end_time = time.time()
duration = (end_time-start_time)/3600 # seconds
if duration<60:
    unit = 'seconds'
    duration = duration*60
else:
    unit = 'minutes'
print(">> elapsed time was %g %s <<" % (duration, unit))
```
